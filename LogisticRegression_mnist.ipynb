```python
# Task1: Extract features for both training set and testing set
# function to find the features - mean and std of input images 
def Calculate_features(file):
    fLength = len(file)
    Features = np.array([None]* fLength)
    for i in range(fLength):
        Features[i]= (np.mean(file[i]), np.std(file[i]))
        
    return Features  
```

```python
import math
```

```python
#define function for gaussian distribution as features are continuous
def calProbability(x,mean,var):
    exponent=math.exp(-math.pow(x-mean,2)/(2*var))
    return((1/math.sqrt(2*math.pi*var))*exponent)
```

```python
# P(x1|class 1) = calProbability(x1, mean_7_x1,var_7_x1)
# P(x2 |Class 1)= CalProbability(x2, mean_7_x2, var_7_x2)
# P(x1|class2)= calProbability(x1, mean_8_x1, var_8_x1)
# P(X2|class2)= calProbability(X2, mean_8_x2, var_8_x2)
#P(Class1) = P(Y=class1)*P(X1|class1)*P(X2|class1)
#P(Class2)= P(Y=class2)*P(X1|class2)*P(X2|class2)
# classifier finds argmax(P(class1), P(class2))
```

```python
# define sigmoid function
# h(x)= g(theta*x)
# z= Theta*x
# g(z)= 1/(1+ exp(-z))
def sigmoid(z):
    return 1/(1+ np.exp(-z))
```

```python
# training the model with Logistic Regression using gradient ascent
def Logistic_Regression(feature_x,learning_rate, epochs, theta, w0):
    trainresults=[]
    while epochs>0:
        trainingprediction=0
        z= w0+ np.dot(feature_x,theta)
        h=sigmoid(z) 
        epochs=epochs-1
        for j in range(len(feature_x)):
            if h[j]>=0.5:
                c_type=1
            else:
                c_type=0
                #compute training accuracy
            if c_type == label_matrix[j]:
                trainingprediction+=1
        #calculate gradient
        gradient_w = np.dot(feature_x.T,(label_matrix-h.T).T)
        gradient_w0 = np.sum(np.dot(feature_x.T, (label_matrix-h.T).T))/len(feature_x)
        #update the training weights
        theta= theta+(learning_rate*gradient_w)
        w0= w0+(learning_rate*gradient_w0) 
        Trainingaccuracy = (trainingprediction/m)*100
        print('\n\nGradient_w:',gradient_w,'Gradient_w0:',gradient_w0, 'Theta:',theta, '\nW0:',w0,'\nTrainingAccuracy:',Trainingaccuracy)

    #training results
    final_theta=theta
    trainresults.append(final_theta)
    final_w0=w0
    trainresults.append(final_w0)
    return trainresults
```

```python
#set the model parameters
learning_rate=0.0005
epochs=10000
m=feature_x.shape[0]
#initialize the weights to random values
theta= np.random.rand(2,1)
w0= np.random.random_sample()
```

```python
train_lr=Logistic_Regression(feature_x,learning_rate, epochs, theta, w0)
```
